```python
import pandas as pd
import scipy.sparse as sp
from sklearn.model_selection import train_test_split
import pickle

print("Libraries loaded successfully")
```

```python
# Load TF-IDF matrix
X = sp.load_npz("/content/drive/MyDrive/FactLens_Group9/data/tfidf_matrix.npz")

# Load labels
y = pd.read_csv("/content/drive/MyDrive/FactLens_Group9/data/labels.csv")["label"]

print(f"Matrix loaded: {X.shape}")
print(f"Labels loaded: {len(y)}")
print(f"\nLabel distribution:")
print(y.value_counts())
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,        # 20% for testing
    random_state=42,      # ensures same split every time
    stratify=y            # keeps label ratio consistent in both splits
)

print("Split complete")
print(f"\nTraining set:  {X_train.shape[0]} articles")
print(f"Test set:      {X_test.shape[0]} articles")
print(f"\nTraining label distribution:")
print(y_train.value_counts())
print(f"\nTest label distribution:")
print(y_test.value_counts())
```

```python
train_pct = X_train.shape[0] / X.shape[0] * 100
test_pct = X_test.shape[0] / X.shape[0] * 100

print(f"Training set:  {train_pct:.1f}%")
print(f"Test set:      {test_pct:.1f}%")

# Verify stratification worked
train_fake_pct = (y_train == "FAKE").sum() / len(y_train) * 100
test_fake_pct = (y_test == "FAKE").sum() / len(y_test) * 100

print(f"\nFake % in training set: {train_fake_pct:.1f}%")
print(f"Fake % in test set:     {test_fake_pct:.1f}%")
print("\nThese two numbers should be very close - that means stratify worked correctly")
```

```python
# Save training and test sets
sp.save_npz("/content/drive/MyDrive/FactLens_Group9/data/X_train.npz", X_train)
sp.save_npz("/content/drive/MyDrive/FactLens_Group9/data/X_test.npz", X_test)

y_train.to_csv("/content/drive/MyDrive/FactLens_Group9/data/y_train.csv", index=False)
y_test.to_csv("/content/drive/MyDrive/FactLens_Group9/data/y_test.csv", index=False)

print("Saved successfully:")
print("  - X_train.npz")
print("  - X_test.npz")
print("  - y_train.csv")
print("  - y_test.csv")
```
